```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('maaslar.csv')
```

```python
df
```

```python
x = df.iloc[:,1:2].values
y = df.iloc[:,2:].values
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=2)
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x,y)
plt.scatter(x,y,color='orange')
plt.plot(x,lr.predict(x),color='black')
```

```python
# degree = 2
pf = PolynomialFeatures(degree=2)
xPoly = pf.fit_transform(x)
lr2 = LinearRegression()
lr2.fit(xPoly,y)
plt.scatter(x,y,color='orange')
plt.plot(x,lr2.predict(xPoly),color='black')
```

```python
# degree = 4
pf = PolynomialFeatures(degree=4)
xPoly = pf.fit_transform(x)
lr2 = LinearRegression()
lr2.fit(xPoly,y)
plt.scatter(x,y,color='orange')
plt.plot(x,lr2.predict(xPoly),color='black')
```

```python
# degree = 8
pf = PolynomialFeatures(degree=8)
xPoly = pf.fit_transform(x)
lr2 = LinearRegression()
lr2.fit(xPoly,y)
plt.scatter(x,y,color='orange')
plt.plot(x,lr2.predict(xPoly),color='black')
```

```python
pf = PolynomialFeatures(degree=10)
xPoly = pf.fit_transform(x)
lr2 = LinearRegression()
lr2.fit(xPoly,y)
plt.scatter(x,y,color='orange')
plt.plot(x,lr2.predict(xPoly),color='black')
```

```python
# predictions

print(lr.predict([[11]]))
print(lr.predict([[6.6]]))
print('------------------')
print(lr2.predict(pf.fit_transform([[11]])))
print(lr2.predict(pf.fit_transform([[6.6]])))
```
