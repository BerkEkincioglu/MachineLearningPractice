```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('musteriler.csv')
```

```python
df.head()
```

```python
plt.scatter(df['Yas'],df['Hacim'],color='orange')
plt.scatter(df['Yas'],df['Maas'],color='black')
plt.show()
```

```python
x = df.iloc[:,3:]
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3,init='k-means++')
kmeans.fit(x)
kmeans.cluster_centers_
```

```python
results = []
for i in range(1,16):
    kmeans_ = KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans_.fit(x)
    results.append(kmeans_.inertia_)
plt.plot(range(1,16),results,color='black')
```
