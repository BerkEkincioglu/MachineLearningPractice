```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('maaslar.csv')
```

```python
df
```

```python
x = df.iloc[:,1:2].values
y = df.iloc[:,2:].values
```

```python
# linear regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x,y)
```

```python
# visualization
plt.scatter(x,y,color='red')
plt.plot(x,lr.predict(x),color='black')
plt.show()
```

```python
# polynomial regression
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=4)
x_poly = pf.fit_transform(x)
lr2= LinearRegression()
lr2.fit(x_poly,y)
# visualization
plt.scatter(x,y,color='red')
plt.plot(x,lr2.predict(pf.fit_transform(x)),color='black')
plt.show()
```

```python
df
```

```python
#predictions
print('Linear Predictions \n')
print(lr.predict([[11]]))
print(lr.predict([[6.6]]))
print(lr.predict([[5]]))
print('Polynomial Predictions \n')
print(lr2.predict(pf.fit_transform([[11]])))
print(lr2.predict(pf.fit_transform([[6.6]])))
print(lr2.predict(pf.fit_transform([[5]])))
```

```python
# verilerin olceklendirilmesi
from sklearn.preprocessing import StandardScaler
sc1 = StandardScaler()
x_olcekli = sc1.fit_transform(x)
sc2 = StandardScaler()
y_olcekli = sc2.fit_transform(y)
```

```python
#SVR
from sklearn.svm import SVR
svr_rbf = SVR('rbf')
svr_rbf.fit(x_olcekli,y_olcekli)
# visualization
plt.scatter(x_olcekli,y_olcekli,color='red')
plt.plot(x_olcekli,svr_rbf.predict(x_olcekli),color='black')
plt.show()
# predictions

print('------------------------')
print(svr_rbf.predict([[11]]))
print(svr_rbf.predict([[6.6]]))
```

```python
svr_lr = SVR('linear')
svr_lr.fit(x_olcekli,y_olcekli)
# visualization
plt.scatter(x_olcekli,y_olcekli,color='red')
plt.plot(x_olcekli,svr_lr.predict(x_olcekli),color='black')
plt.show()
```

```python
svr_poly = SVR('poly')
svr_poly.fit(x_olcekli,y_olcekli)
# visualization
plt.scatter(x_olcekli,y_olcekli,color='red')
plt.plot(x_olcekli,svr_poly.predict(x_olcekli),color='black')
plt.show()
```

```python
svr_sgmd = SVR('sigmoid')
svr_sgmd.fit(x_olcekli,y_olcekli)
# visualization
plt.scatter(x_olcekli,y_olcekli,color='red')
plt.plot(x_olcekli,svr_sgmd.predict(x_olcekli),color='black')
plt.show()
```
