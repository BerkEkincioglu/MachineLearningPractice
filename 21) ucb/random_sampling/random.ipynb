```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import random
```

```python
df = pd.read_csv('Ads_CTR_Optimisation.csv')
```

```python
N = 10000
d = 10
summary = 0
selected = []
for n in range(0,N):
    ad = random.randrange(d)
    selected.append(ad)
    award = df.values[n,ad]
    summary+=award
plt.hist(selected,color='black')
plt.show()
summary
```
